# scraping website

```python
url="https://myfirstdrone.com/drones-for-sale"
```

# response

```python
response=requests.get(url)
```

```python
response
```

# parsing

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
drone=soup.select(".brand+ .drone-name")
```

```python
drone
```

```python
drone_list=[]
```

```python
for item in drone:
    drone_list.append(item.text)
    drone_list
```

```python
drone_list
```

```python
rating=soup.select('.small-rating-circle:nth-child(1)')
```

```python
rating
```

```python
value_rating=[]
```

```python
for item in rating:
    value_rating.append(item.text)
    value_rating
```

```python
value_rating
```

```python
flying=soup.select('.features-left-column .column-10 .inline:nth-child(1)')
```

```python
flying
```

```python
flying_time
```
